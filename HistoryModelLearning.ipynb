```python
import pandas as pd
import matplotlib.pyplot as plt
import pathlib
```

```python
day = '2024-01-05'
```

```python
pathlib.Path('./Picture/V1-'+day).mkdir(parents=True, exist_ok=True) 
```

```python
modelLearning = pd.read_csv('./Results/'+day+'/HistoryModelLearning-'+day+'.csv')
modelLearningLeNet5 = pd.read_csv('./Results/'+day+'/HistoryModelLearningLeNet5-'+day+'.csv')
modelLearningAlexNet = pd.read_csv('./Results/'+day+'/HistoryModelLearningAlexNet-'+day+'.csv')
modelLearningMobileNet = pd.read_csv('./Results/'+day+'/HistoryModelLearningMobileNet-'+day+'.csv')
```

```python
plt.title('Validation Accuracy')
plt.plot(modelLearning['val_accuracy'],label='DCC')
plt.plot(modelLearningLeNet5['val_accuracy'],label='LeNet5')
plt.plot(modelLearningAlexNet['val_accuracy'],label='AlexNet')
plt.plot(modelLearningMobileNet['val_accuracy'],label='MobileNet')
plt.legend()
plt.savefig('./Picture/V1-'+day+'/HistoryModelLearning-ValidationAccuracy-'+day+'.jpeg', dpi = 900)
plt.show()
```

```python
plt.title('Accuracy')
plt.plot(modelLearning['accuracy'],label='DCC')
plt.plot(modelLearningLeNet5['accuracy'],label='LeNet5')
plt.plot(modelLearningAlexNet['accuracy'],label='AlexNet')
plt.plot(modelLearningMobileNet['accuracy'],label='MobileNet')
plt.legend()
plt.savefig('./Picture/V1-'+day+'/HistoryModelLearning-Accuracy-'+day+'.jpeg', dpi = 900)
plt.show()
```

```python
plt.title('Validation Loss')
plt.plot(modelLearning['val_loss'],label='DCC')
plt.plot(modelLearningLeNet5['val_loss'],label='LeNet5')
plt.plot(modelLearningAlexNet['val_loss'],label='AlexNet')
plt.plot(modelLearningMobileNet['val_loss'],label='MobileNet')
plt.ylim(0,3)
plt.legend()
plt.savefig('./Picture/V1-'+day+'/HistoryModelLearning-ValidationLoss-'+day+'.jpeg', dpi = 900)
plt.show()
```

```python
plt.title('Loss')
plt.plot(modelLearning['loss'],label='DCC')
plt.plot(modelLearningLeNet5['loss'],label='LeNet5')
plt.plot(modelLearningAlexNet['loss'],label='AlexNet')
plt.plot(modelLearningMobileNet['loss'],label='MobileNet')
plt.legend()
plt.savefig('./Picture/V1-'+day+'/HistoryModelLearning-Loss-'+day+'.jpeg', dpi = 900)
plt.show()
```

```python
accuracy = pd.DataFrame()
accuracy['DCC'] = modelLearning['accuracy']
accuracy['LeNet5'] = modelLearningLeNet5['accuracy']
accuracy['AlexNet'] = modelLearningAlexNet['accuracy']
accuracy['MobileNet']=modelLearningMobileNet['accuracy']
```

```python
val_accuracy = pd.DataFrame()
val_accuracy['DCC'] = modelLearning['val_accuracy']
val_accuracy['LeNet5'] = modelLearningLeNet5['val_accuracy']
val_accuracy['AlexNet'] = modelLearningAlexNet['val_accuracy']
val_accuracy['MobileNet']=modelLearningMobileNet['val_accuracy']
```

```python
loss = pd.DataFrame()
loss['DCC'] = modelLearning['loss']
loss['LeNet5'] = modelLearningLeNet5['loss']
loss['AlexNet'] = modelLearningAlexNet['loss']
loss['MobileNet']= modelLearningMobileNet['loss']
```

```python
val_loss = pd.DataFrame()
val_loss['DCC'] = modelLearning['val_loss']
val_loss['LeNet5'] = modelLearningLeNet5['val_loss']
val_loss['AlexNet'] = modelLearningAlexNet['val_loss']
val_loss['MobileNet'] =modelLearningMobileNet['val_loss']
```

```python
fig, axes = plt.subplots(2,2, sharex=True, sharey=True, )
fig_accuracy = accuracy.plot(ax=axes[0][0])
fig_accuracy.set_title('Accuracy')
fig_accuracy.legend()
fig_val_accuracy = val_accuracy.plot(ax=axes[0][1])
fig_val_accuracy.set_title('Validation Accuracy')
fig_val_accuracy.legend()
fig_loss = loss.plot(ax=axes[1][0])
fig_loss.set_title('Loss')
fig_loss.legend()
fig_val_loss = val_loss.plot(ax=axes[1][1])
fig_val_loss.set_title('Validation Loss')
fig_val_loss.legend()
fig.savefig('./Picture/V1-'+day+'/HistoryModelLearning-'+day+'.jpeg', dpi = 900)
plt.ylim(0,2)
plt.show()
```
